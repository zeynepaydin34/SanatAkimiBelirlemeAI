```python
import os
import imagehash
from PIL import Image
from collections import defaultdict

# Görüntünün pHash değerini hesaplama
def calculate_phash(image_path):
    image = Image.open(image_path)
    return imagehash.phash(image)

# Aynı olan görüntüleri bulma ve silme
def find_and_remove_duplicates(base_folder):
    hash_dict = defaultdict(list)
    
    # Tüm klasörleri ve görüntü dosyalarını tarama
    for root, dirs, files in os.walk(base_folder):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                file_path = os.path.join(root, file)
                image_hash = calculate_phash(file_path)
                hash_dict[image_hash].append(file_path)
    
    # Aynı olan görüntüleri bulma ve birini silme
    deleted_files = []
    for image_hash, paths in hash_dict.items():
        if len(paths) > 1:
            print(f"Aynı görüntüler: {paths}")
            # İlk resmi tutup diğerlerini silme
            for duplicate_image in paths[1:]:
                os.remove(duplicate_image)  # Duplicate olan resmi silme
                deleted_files.append(duplicate_image)
                print(f"Silinen dosya: {duplicate_image}")
    
    return deleted_files

base_folder = r"E:\YazLab Proje\Impressionism"  
deleted_images = find_and_remove_duplicates(base_folder)

# Silinen resimlerin listesi ve sayısı
print("\nSilinen Görüntüler:")
for image in deleted_images:
    print(image)

print(f"\nToplam Silinen Resim Sayısı: {len(deleted_images)}")
```
